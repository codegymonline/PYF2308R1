```python
myfile = open("workinghours.txt", encoding="utf8")

# Đọc file
lines = myfile.readlines()
print(lines)
myfile.close()

for line in lines:
    # print(line)
    data = line.split()
    print("Đọc được và phân tách:", data)
    hours = list(map(lambda item: float(item), data[2:]))
    print(data[0], data[1], sum(hours))

    # total_hours = 0
    # for item in data[2:]:
    #     total_hours += float(item)
```

```python
# Demo 2 - Nhập tên file, nhập nội dung rồi ghi xuống file với tên vừa nhập
ten_file = input("Nhập tên file:").strip()
if ten_file != "":
    noi_dung = input("Nhập nội dung:")
    with open(ten_file, "a", encoding="utf8") as myfile:
        myfile.write(noi_dung)
```

```python
# Ghi file json
import json
data = [
    {
        "ma": "HV001",
        "ten": "Lý Tiểu Mẫn",
        "diem": 9.7,
        "gioi_tinh": True
    },
    {
        "ma": "HV002",
        "ten": "Lý Tiểu Long",
        "diem": 8,
        "gioi_tinh": False
    }
]
with open("hocvien.json", "w", encoding="utf8") as myfile:
    json.dump(data, myfile, indent=4)
```

```python
# Demo call API
import requests
import json

r = requests.get("https://dummyjson.com/products")
if r.status_code == 200: # Success
    data = []
    for item in r.json()["products"]:
        print(item["id"], item["title"], item["price"], item["brand"])
        data.append({
            "masp": item["id"],
            "tensp": item["title"],
            "dongia": item["price"],
            "loai": item["brand"]
        })

    print(data)
    with open("products.json", "w", encoding="utf8") as file:
        json.dump(data, file, indent=4)
```
